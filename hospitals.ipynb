```python
import utils
import os
import re
import shutil
from tqdm import tqdm
```

```python
directory = "/local/data1/jakli758/accepted_anonymized_hospitalinfo_8bit_Sept15_2023"
for filename in tqdm(os.listdir(directory)):
        if not filename.endswith(".png"):
            continue

        # Extract hospital number
        hospital_match = re.search(r"hospital_(\d+)", filename)

        # Extract label
        label_match = re.search(r"_(AFF|NFF)_", filename)

        if hospital_match and label_match:
            hospital = int(hospital_match.group(1))
            label = label_match.group(1)
            path = os.path.join(directory, str(hospital), str(label))
            os.makedirs(path, exist_ok=True)
            shutil.copy(os.path.join(directory, filename), os.path.join(path, filename))
            
```
