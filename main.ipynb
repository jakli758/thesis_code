```python
import os
def test_preprocess(path='/local/data1/jakli758/threeclasses/', new_shape=(128,128)):

    folders = ['accepted', 'rejected']
    
    for folder in folders:
        path_current = f'{path}{folder}'
        path_new = f'{path}{folder}_preprocessed'
        for file in os.listdir(path_current):
            

            # imread automatically converts img to 8-bit, if CV.IMREAD_ANYDEPTH is not set
            img = cv2.imread(f'{path_current}/{file}', 0)
            img_preprocessed = preprocessing.resize_with_pad(image=img,new_shape=new_shape, padding_color=0)
            plt.imshow(img_preprocessed, cmap='gray')
            print(f'{path_current}/{file}')
            print(f'{path_new}/{file}')
            #cv2.imwrite(f'{path_new}/{file}')
            break
        break
```
